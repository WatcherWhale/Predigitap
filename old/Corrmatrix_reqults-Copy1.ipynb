```python
dataset = "WT_2"
title = "Web Technology (2021-2022)"

input, output = fullDataSet("./datasets/raw/logs_" + dataset + ".csv", "./datasets/raw/Resultaten_" + dataset + ".csv", None, datetime.datetime(2023, 2, 10, 0, 0).timestamp())
saveDataSet(input, output, "./datasets")
```

```python
data = pd.read_csv("./datasets/input.csv", sep=';')
#data = data.loc[:, (data != 0).any(axis=0)]
data
```

```python
comatrix = data.corr()
res = comatrix["results"]
labels = list(res._mgr.axes[0])
res
```

```python
plt.title("Correlatie " + title)
sn.heatmap(comatrix)
```

```python
plt.tight_layout()
plt.title("Correlatie " + title)
sn.heatmap(np.array(res).reshape((14,1)), annot=True, yticklabels=labels, xticklabels=[])
plt.savefig("corrmatrix/" + dataset + ".png", dpi=400, bbox_inches='tight')
```
